## Description
**StudentID**: A unique identifier assigned to each student (1001 to 3392).
### Demographic Details
**Age:** The age of the students ranges from 15 to 18 years.
**Gender:** Gender of the students, where 0 represents Male and 1 represents Female.
**Ethnicity:** The ethnicity of the students, coded as follows:
- 0: Caucasian
- 1: African American
- 2: Asian
- 3: Other
**ParentalEducation:** The education level of the parents, coded as follows:
- 0: None
- 1: High School
- 2: Some College
- 3: Bachelor's
- 4: Higher
### Study Habits
**StudyTimeWeekly:** Weekly study time in hours, ranging from 0 to 20.
**Absences:** Number of absences during the school year, ranging from 0 to 30.
**Tutoring:** Tutoring status, where 0 indicates No and 1 indicates Yes.
### Parental Involvement
**ParentalSupport:** The level of parental support, coded as follows:
- 0: None
- 1: Low
- 2: Moderate
- 3: High
- 4: Very High
### Extracurricular Activities
**Extracurricular:** Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.
**Sports:** Participation in sports, where 0 indicates No and 1 indicates Yes.
**Music:** Participation in music activities, where 0 indicates No and 1 indicates Yes.
**Volunteering:** Participation in volunteering, where 0 indicates No and 1 indicates Yes.
### Academic Performance
**GPA**: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities**.
**Target** Variable: Grade Class
**GradeClass:** Classification of students' grades based on GPA:
- 0: 'A' (GPA >= 3.5)
- 1: 'B' (3.0 <= GPA < 3.5)
- 2: 'C' (2.5 <= GPA < 3.0)
- 3: 'D' (2.0 <= GPA < 2.5)
- 4: 'F' (GPA < 2.0)

```python
# Replacing the float numbers to conventional letter grading 
formatted_data = data.replace({"GradeClass":{0.0:"A",1.0:"B",2.0:"C",3.0:"D",4.0:"F"}},inplace=False)
formatted_data
```

```python
# Converted to category type for memory, cpu efficiency, enforcing data integrity and facilitating data analysis easily 
formatted_data["GradeClass"].astype("category")
```

```python
# Defining function to normalize the features (feature scaling)
def z_score_norm(value,mean,standard_deviation):
    return (value-mean)/standard_deviation
```

```python
# Get necessary data for column
std_study = formatted_data["StudyTimeWeekly"].std()
mean_study = formatted_data["StudyTimeWeekly"].mean()
std_absence = formatted_data["Absences"].std()
mean_absence = formatted_data["Absences"].mean()
```

```python
# Apply the normalization to the studyTimeWeekly column
formatted_data["StudyTimeWeekly"] = z_score_norm(formatted_data["StudyTimeWeekly"],mean_study,std_study)
formatted_data["Absences"] = z_score_norm(formatted_data["Absences"],mean_absence,std_absence)
```

```python
semi_data = data
semi_data = semi_data.replace({"GradeClass":{0.0:"A",1.0:"B",2.0:"C",3.0:"D",4.0:"F"}},inplace=False)
semi_data["GradeClass"].astype("category")
semi_data = pd.get_dummies(semi_data,"GradeClass")
semi_x = semi_data.drop(['StudentID','GradeClass_A','GradeClass_B','GradeClass_C','GradeClass_D','GradeClass_F','GPA'],axis=1)
semi_y = semi_data[['GradeClass_A','GradeClass_B','GradeClass_C','GradeClass_D','GradeClass_F']]
semi_y
```

```python
def init_data(data):
    """
    initializes data.
 
    Parameters
    ----------
    data : pandas.core.frame.DataFrame
        the raw data of the model containing:StudentID, Age, Gender, Ethnicity, ParentalEducation, StudyTimeWeekly, Absences, Tutoring, ParentalSupport, Extracurricular, Sports, Music, Volunteering, GPA, GradeClass
 
    Returns
    -------
    pandas.core.frame.DataFrame
        The data with replaced gradeclass as letter grading and added ParentalEffect column
    """
    temp = data
    temp['GradeClass'] = temp['GPA'].apply(assign_grade_class)
    temp = temp.replace({"GradeClass":{0.0:"A",1.0:"B",2.0:"C",3.0:"D",4.0:"F"}},inplace=False)
    temp["ParentalEffect"] = temp["ParentalSupport"] * temp["ParentalEducation"]
    return temp
    
def divide_data(data):
    x = data.drop(['StudentID','GradeClass_A','GradeClass_B','GradeClass_C','GradeClass_D','GradeClass_F','GPA','Gender','Ethnicity'],axis=1)
    y = data[['GradeClass_A','GradeClass_B','GradeClass_C','GradeClass_D','GradeClass_F']]
    return x,y
    
def normalize_data(data):
    temp = data
    std_pe = temp["ParentalEffect"].std()
    mean_pe = temp["ParentalEffect"].mean()
    temp["ParentalEffect"] = z_score_norm(temp["ParentalEffect"],mean_pe,std_pe)
    temp["Absences"] = z_score_norm(temp["Absences"],temp["Absences"].mean(),temp["Absences"].std())
    temp["ParentalEducation"] = z_score_norm(temp["ParentalEducation"],temp["ParentalEducation"].mean(),temp["ParentalEducation"].std())
    temp["StudyTimeWeekly"] = z_score_norm(temp["StudyTimeWeekly"],temp["StudyTimeWeekly"].mean(),temp["StudyTimeWeekly"].std())
    temp["ParentalSupport"] = z_score_norm(temp["ParentalSupport"],temp["ParentalSupport"].mean(),temp["ParentalSupport"].std())
    return temp

def one_hot_encoding(data):
    temp = data
    categories = ['A', 'B', 'C', 'D', 'F']
    temp['GradeClass'] = pd.Categorical(temp['GradeClass'], categories=categories)
    temp = pd.get_dummies(temp,columns=['GradeClass'])
    return temp

def format_data(data):
    temp = init_data(data)
    temp = one_hot_encoding(temp)
    temp = normalize_data(temp)
    temp = divide_data(temp)
    return temp

def create_random_raw_data(num_samples):
    test_data = pd.DataFrame({
        'StudentID': np.arange(1, num_samples+1),
        'Age': np.random.randint(15, 18, num_samples),
        'Gender': np.random.choice([0, 1], num_samples),
        'Ethnicity': np.random.choice(['0', '1', '2','3'], num_samples),
        'ParentalEducation': np.random.choice([0, 1, 2, 3, 4], num_samples),
        'StudyTimeWeekly': np.random.uniform(1, 20, num_samples),
        'Absences': np.random.randint(0, 30, num_samples),
        'Tutoring': np.random.choice([0, 1], num_samples),
        'ParentalSupport': np.random.choice([0, 1, 2, 3, 4], num_samples),
        'Extracurricular': np.random.choice([0, 1], num_samples),
        'Sports': np.random.choice([0, 1], num_samples),
        'Music': np.random.choice([0, 1], num_samples),
        'Volunteering': np.random.choice([0, 1], num_samples),
        'GPA': np.round(np.random.uniform(2.0, 4.0, num_samples), 2),
        'GradeClass': np.random.choice([0.0,1.0,2.0,3.0,4.0], num_samples)
    })
    return test_data

def create_data(student_id,age,gender,ethnicity,parental_education,study_time_weekly,absences,tutoring,parental_support,extracurricular,sports,music,volunteering,gpa,grade_class):
    data = pd.DataFrame({
        'StudentID': student_id,
        'Age': age,
        'Gender': gender,
        'Ethnicity': ethnicity,
        'ParentalEducation': parental_education,
        'StudyTimeWeekly': study_time_weekly,
        'Absences': absences,
        'Tutoring': tutoring,
        'ParentalSupport': parental_support,
        'Extracurricular': extracurricular,
        'Sports': sports,
        'Music': music,
        'Volunteering': volunteering,
        'GPA': gpa,
        'GradeClass': grade_class
    })
    return data

def predict_gpa(model,data):
    x,y = format_data(data)
    predictions = model.predict(x)
    predicted_gpa = prediction_to_grade(predictions)
    y_compare = (data.replace({"GradeClass":{0.0:"A",1.0:"B",2.0:"C",3.0:"D",4.0:"F"}},inplace=False))["GradeClass"]
    output = pd.concat([x,y_compare ,predicted_gpa], axis=1)
    return output

def prediction_to_binary(predictions):
    predicted_indices = np.argmax(predictions, axis=1)
    categories = ['GradeClassA', 'GradeClassB', 'GradeClassC', 'GradeClassD', 'GradeClassF']
    
    # Initialize a DataFrame with zeros for each category
    binary_columns = pd.DataFrame(0, index=np.arange(len(predicted_indices)), columns=categories)
    
    # Set the appropriate category column to 1 based on the prediction
    for i, idx in enumerate(predicted_indices):
        binary_columns.loc[i, categories[idx]] = 1
    return binary_columns

def prediction_to_grade(predictions):
    predicted_categories = np.argmax(predictions, axis=1)
    # Extract the predicted category indices
    index_to_category = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'F'}
    
    predicted_gpa = [index_to_category[idx] for idx in predicted_categories]
    return pd.DataFrame({"PredictedGpa":predicted_gpa})

def add_grade_category(data,category,value,multiple):
    data_category = data[data[category]==value]
    data_added = data
    for i in range(multiple):
        data_added = pd.concat([data_added,data_category],ignore_index=True)
    return data_added

def create_custom_grade_class_data(num_samples, grade_class_value):
    # Generate synthetic data
    data = {
        'StudentID': np.arange(1, num_samples + 1),
        'Age': np.random.randint(15, 18, num_samples),
        'Gender': np.random.choice([0, 1], num_samples),  # 0 for female, 1 for male
        'Ethnicity': np.random.choice([0, 1, 2, 3], num_samples),
        'ParentalEducation': np.random.choice([0, 1, 2, 3, 4], num_samples),
        'StudyTimeWeekly': np.random.uniform(19, 20, num_samples),  # Higher study time for better grades
        'Absences': np.random.randint(0, 3, num_samples),  # Lower absences for better grades
        'Tutoring': np.random.choice([0, 1], num_samples),
        'ParentalSupport': np.random.choice([0, 1, 2, 3, 4], num_samples),  # Higher parental support for better grades
        'Extracurricular': np.random.choice([0, 1], num_samples),
        'Sports': np.random.choice([0, 1], num_samples),
        'Music': np.random.choice([0, 1], num_samples),
        'Volunteering': np.random.choice([0, 1], num_samples),
        'GPA': np.round(np.random.uniform(3.5, 4.0, num_samples), 2),  # Higher GPA for GradeClass A
        'GradeClass': grade_class_value
    }

    # Create DataFrame
    custom_data = pd.DataFrame(data)
    
    return custom_data


def assign_grade_class(gpa):
    if gpa >= 3.5:
        return 'A'
    elif 3.0 <= gpa < 3.5:
        return 'B'
    elif 2.5 <= gpa < 3.0:
        return 'C'
    elif 2.0 <= gpa < 2.5:
        return 'D'
    else:
        return 'F'
```
